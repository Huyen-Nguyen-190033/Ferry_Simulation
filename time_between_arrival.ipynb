```python
# Function to convert timestamps to minutes
def convert_to_minutes(timestamps):
    minutes_list = []
    for ts in timestamps:
        h, m, s = map(int, ts.split(":"))
        total_minutes = h * 60 + m + s / 60
        minutes_list.append(total_minutes)
    return minutes_list

# Convert each list to minutes
timestamps1_minutes = convert_to_minutes(timestamps1)
timestamps2_minutes = convert_to_minutes(timestamps2)
timestamps3_minutes = convert_to_minutes(timestamps3)

timestamps1_minutes, timestamps2_minutes, timestamps3_minutes
```

```python
import matplotlib.pyplot as plt

# Function to plot the timestamps with fixed x-bound
def plot_timestamps_fixed_x(timestamps, title):
    x = timestamps  # Use the timestamps in minutes as x-axis
    y = range(len(timestamps))  # Increment y for each timestamp

    plt.figure(figsize=(10, 6))
    # Plot the main line excluding the final value
    plt.plot(x, y, marker='o', linestyle='-', label=title, color="#7eb0d5")
    # Add a scatter point to highlight the final value
    plt.scatter(x[-1], y[-1], color="#fd7f6f", zorder=5, label=f"Final Value ({x[-1]:.2f}, {y[-1]})")

    plt.title(f"Vehicle Arrivals over time for {title}")
    plt.xlabel("Timestamps (Minutes)")
    plt.ylabel("Number of Vehicles")
    plt.xlim(0, 30)
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage (timestamps1_minutes, timestamps2_minutes, timestamps3_minutes need to be defined)
plot_timestamps_fixed_x(timestamps1_minutes, "Timestamps1")
plot_timestamps_fixed_x(timestamps2_minutes, "Timestamps2")
plot_timestamps_fixed_x(timestamps3_minutes, "Timestamps3")
```

```python
import numpy as np
import pandas as pd


# Calculate time between arrivals
time_between_arrivals = np.diff(arrival_times)

# Create a DataFrame for better visualization
df_arrivals = pd.DataFrame({
    "Arrival Time": arrival_times[:-1],
    "Next Arrival Time": arrival_times[1:],
    "Time Between Arrivals": time_between_arrivals
})

df_arrivals
```

```python
# Plotting the line graph for time between arrivals
plt.figure(figsize=(8, 5))
plt.plot(df_arrivals.index, df_arrivals["Time Between Arrivals"], marker='o', linestyle='-', label='Time Between Arrivals')
plt.title('Time Between Arrivals')
plt.xlabel('Index')
plt.ylabel('Time Between Arrivals (minutes)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from scipy.stats import expon

# Fit the time between arrivals into an exponential distribution
params = expon.fit(time_between_arrivals, floc=0)  # floc=0 ensures fit starts at zero
mean_time_between_arrivals = params[1]

mean_time_between_arrivals
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon

# Generate x values for the exponential distribution
x = np.linspace(0, max(time_between_arrivals), 1000)
y = expon.pdf(x, scale=mean_time_between_arrivals)

# Plot the exponential distribution
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Exponential Distribution (Mean = {:.3f})'.format(mean_time_between_arrivals), color = "#fd7f6f")
plt.hist(time_between_arrivals, bins=30, density=True, alpha=0.6, color='#7eb0d5', label='Observed Data')
plt.title('Fitted Exponential Distribution for Time Between Arrivals')
plt.xlabel('Time Between Arrivals (minutes)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import norm

# Calculate the cumulative sum of time between arrivals to simulate arrival times
cumulative_arrival_times = np.cumsum(time_between_arrivals)

# Estimate arrival time for the 45th car
mean_arrival_time_45th = cumulative_arrival_times[44]  # Mean arrival time for the 45th car
std_dev_time_between = np.std(time_between_arrivals, ddof=1)  # Standard deviation of time between arrivals
std_error_45th = std_dev_time_between * np.sqrt(45)  # Standard error for the sum of 45 intervals

# Use the normal distribution to calculate the 99% confidence interval for the 45th car
z_value = norm.ppf(0.95)  # Z-value for 99% confidence
lower_bound = mean_arrival_time_45th - z_value * std_error_45th
upper_bound = mean_arrival_time_45th + z_value * std_error_45th

mean_arrival_time_45th, lower_bound, upper_bound
```

```python
# Generate x values for the normal distribution
x = np.linspace(lower_bound - 1, upper_bound + 1, 1000)
y = norm.pdf(x, loc=mean_arrival_time_45th, scale=std_error_45th)

# Plot the normal distribution
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Normal Distribution (Mean = {:.2f})'.format(mean_arrival_time_45th))
# plt.axvline(lower_bound, color='red', linestyle='--', label='99% Confidence Interval Lower Bound')
# plt.axvline(upper_bound, color='red', linestyle='--', label='99% Confidence Interval Upper Bound')
plt.axvline(mean_arrival_time_45th, color='blue', linestyle='-', label='Mean Arrival Time')
plt.fill_between(x, y, where=(x >= lower_bound) & (x <= upper_bound), color='lightgray', alpha=0.5, label='95% Confidence Interval')
plt.title('Normal Distribution for Arrival Time of 45th Car')
plt.xlabel('Time (minutes)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
