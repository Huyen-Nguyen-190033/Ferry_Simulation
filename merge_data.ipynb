```python
import pandas as pd
import numpy as np
import os

```

```python

# Specify the folder containing the CSV files
folder_path = '/content/drive/MyDrive/Sim_FinalProject_FerryTraffic/Huyen/Clean_Data/Scotland_0621_0824/Scotland_21'

# List to hold each DataFrame
dfs = []

# Loop through each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Read the CSV file into a DataFrame and add it to the list
        file_path = os.path.join(folder_path, file_name)
        dfs.append(pd.read_csv(file_path))

# Concatenate all DataFrames in the list into a single DataFrame
merged_df_21 = pd.concat(dfs, ignore_index=True)

# Display the merged DataFrame
merged_df_21
```

```python
# Count occurrences of each date
date_counts = merged_df_js['Date'].value_counts()

# Filter dates that appear more than 60 times
dates_more_than_60 = date_counts[date_counts > 55]

# Display the results
dates_more_than_60
```
