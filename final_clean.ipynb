```python
import pandas as pd
import numpy as np

merged_df_sc = pd.read_csv('/content/drive/MyDrive/Sim_FinalProject_FerryTraffic/Huyen/Clean_Data/Jamestown_0621_0824/js_total.csv')
merged_df_sc = pd.read_csv('/content/drive/MyDrive/Sim_FinalProject_FerryTraffic/Huyen/Clean_Data/Scotland_0621_0824/sc_total.csv')

# Ensure the 'date' column is in datetime format
merged_df_sc['Date'] = pd.to_datetime(merged_df_sc['Date'])

# Define the start and end dates for the full date range
start_date = '2021-06-01'
end_date = '2024-08-31'

# Generate the complete date range from June 2021 to August 2024
full_date_range = pd.date_range(start=start_date, end=end_date)

# Get unique dates present in the DataFrame
dates_in_df = merged_df_sc['Date'].unique()

# Find missing dates by checking for dates in full_date_range not in dates_in_df
missing_dates = full_date_range[~full_date_range.isin(dates_in_df)]

# Check if there are any missing dates
if len(missing_dates) > 0:
    print("Missing dates found:")
    print(missing_dates)
else:
    print("All dates from June 2021 to August 2024 are present.")
```

```python
file_path = '/content/drive/MyDrive/Sim_FinalProject_FerryTraffic/Data/WM data/Traffic 2023/12 TRAFFIC Dec 2023.xlsm'
sheet_name = 'HolidayCalendar'

df_holiday = pd.read_excel(file_path, sheet_name=sheet_name)[['Year', 'Holiday', 'Holiday Name']]
```

```python
# Ensure both date columns are in datetime format for accurate matching
merged_df_sc['Date'] = pd.to_datetime(merged_df_sc['Date'])
df_holiday['Holiday'] = pd.to_datetime(df_holiday['Holiday'])

# Merge the DataFrames on the date columns, using a left join to keep all dates in merged_df_sc
merged_df_sc = merged_df_sc.merge(df_holiday[['Holiday', 'Holiday Name']],
                                  how='left',
                                  left_on='Date',
                                  right_on='Holiday')

# Rename the "Holiday Name" column to "Holiday" in the merged DataFrame
merged_df_sc['Holiday'] = merged_df_sc['Holiday Name']

# Drop the extra "Holiday" column from df_holiday that was used for the merge
merged_df_sc = merged_df_sc.drop(columns=['Holiday_x', 'Holiday Name'], errors='ignore')

# Display the updated DataFrame
merged_df_sc
```

```python
# Filter the DataFrame for rows where the 'Holiday' column is not NaN
holidays_in_df = merged_df_sc[merged_df_sc['Holiday'].notna()]

# Display the rows with non-NaN values in the 'Holiday' column
holidays_in_df
```

```python
import pandas as pd

# Ensure 'Date' is in datetime format if not already
merged_df_sc['Date'] = pd.to_datetime(merged_df_sc['Date'])

# Define mappings for Shift and Hour columns
shift_mapping = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5}
hour_mapping = {'Slot 1': 1, 'Slot 2': 2, 'Slot 3': 3, 'Slot 4': 4, 'Slot 5': 5,
                'Slot 6': 6, 'Slot 7': 7, 'Slot 8': 8, 'Slot 9': 9, 'Slot 10': 10,
                'Slot 11': 11, 'Slot 12': 12}

# Map Shift and Hour columns to their respective numerical order
merged_df_sc['Shift_order'] = merged_df_sc['Shift'].map(shift_mapping)
merged_df_sc['Hour_order'] = merged_df_sc['Hour'].map(hour_mapping)

# Sort the DataFrame by Date, Shift_order, and Hour_order
merged_df_sc = merged_df_sc.sort_values(
    by=['Date', 'Shift_order', 'Hour_order'],
    ascending=[True, True, True]
)

# Drop the temporary ordering columns
merged_df_sc = merged_df_sc.drop(columns=['Shift_order', 'Hour_order'])

# Display the sorted DataFrame
merged_df_sc
```
