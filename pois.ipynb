```python
from scipy.optimize import minimize_scalar
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson


# Function to find the mean value given the 99th percentile (upper_tail_99)
target_upper_tail_99 = 44

def objective(mean_value):
    return (poisson.ppf(0.99, mean_value) - target_upper_tail_99) ** 2

# Optimize to find the mean value
result = minimize_scalar(objective, bounds=(1, 100), method='bounded')

result.x  # Estimated mean value
```

```python
# Parameters for the Poisson distribution
estimated_mean = result.x
x_values = np.arange(0, estimated_mean + estimated_mean)  # Extended range for visibility

# Poisson PMF
pmf_values = poisson.pmf(x_values, estimated_mean)

# Plot the Poisson distribution
plt.figure(figsize=(10, 6))
plt.stem(x_values, pmf_values, linefmt='#7eb0d5', markerfmt='#ffb55a', basefmt=' ', label=f'Estimated Mean = {estimated_mean:.2f}')
plt.axvline(x=target_upper_tail_99, color='#fd7f6f', linestyle='--', label=f'99th Percentile = {target_upper_tail_99}')
plt.title('Poisson Distribution for Small Ferries')
plt.xlabel('Vehicle Counts')
plt.ylabel('Probability')
plt.legend()
plt.grid(alpha=0.5)
plt.show()

```
