```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

total_2_values = lod_positive['TOTAL_2']

# Fit a normal distribution to the data
mu, std = norm.fit(total_2_values)

# Generate data for plotting the fitted distribution
x = np.linspace(total_2_values.min(), total_2_values.max(), 1000)
pdf = norm.pdf(x, mu, std)

# Plot the histogram and fitted distribution
plt.figure(figsize=(10, 6))
plt.hist(total_2_values, bins=20, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram')
plt.plot(x, pdf, 'r-', label=f'Normal Dist.\nMean = {mu:.2f}, Std = {std:.2f}')
plt.title('Fitted Distribution to TOTAL_2 when LOD > 0')
plt.xlabel('TOTAL_2')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

(mu, std)
```

```python
from scipy.stats import lognorm

# Fit a log-normal distribution to the data
shape, loc, scale = lognorm.fit(total_2_values, floc=0)

# Generate data for plotting the fitted log-normal distribution
x = np.linspace(total_2_values.min(), total_2_values.max(), 1000)
pdf = lognorm.pdf(x, shape, loc, scale)

# Plot the histogram and fitted log-normal distribution
plt.figure(figsize=(10, 6))
plt.hist(total_2_values, bins=20, density=True, alpha=0.6, edgecolor='black', label='Distribution Histogram', color = "#7eb0d5")
plt.plot(x, pdf, 'b-', label=f'Log-Normal Dist.\nShape = {shape:.2f}, Scale = {scale:.2f}', color = "#fd7f6f")
plt.title('Log-Normal Distribution of Total Number of Vehicles in 30-min cycles for Big Ferries')
plt.xlabel('Number of Vehicles')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

(shape, loc, scale)
```
